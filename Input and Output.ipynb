## Export DataFrame to CSV File with .to_csv() Method

```python
#this is for when data is updated on a daily basis online
URL = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv?accessType=DOWNLOAD&bom=true&format=true&delimiter=%3B"
baby_names = pd.read_csv(URL, sep=";")
baby_names.head(3)
```

```python
#to save the modified dataframe as an excel file; use True for index if you want the indeces to be also written
baby_names.to_csv("NYC Baby Names.csv", index = False)
```

```python
#write only a certain columns, not the entire dataframe to a csv file
baby_names.to_csv("NYC Baby Names 2.csv", index=False, columns=["Year of Birth", "Gender", "Child's First Name", "Count"])
```

```python
#sometimes when working with foreign characters there will be encoding errors, in this case:
baby_names.to_csv("NYC Baby Names 3.csv", index = False, columns=["Gender", "Rank", "Count"], encoding="utf-8")
```

## Import Excel File

```python
#import xlsx files using the .read_excel extention
pd.read_excel("Data - Single Worksheet.xlsx")
```

```python
#if the excel file has multiple sheets, there is a sheetname parameter; this parameter can also accept a list
pd.read_excel("Data - Multiple Worksheets.xlsx", sheetname = 1)
```

```python
#read multiple sheets using the sheet and a list parameter; the data here will be a dictionary after the import 
data = pd.read_excel("Data - Multiple Worksheets.xlsx", sheetname=[0, 1])
```

```python
#proof
type(data)
```

```python
#we can also automate the process using the None keyword for the sheetname parameter
data = pd.read_excel("Data - Multiple Worksheets.xlsx", sheetname=None)
data
```

```python
#the type is still the same
type(data)
```

## Export Excel File

```python
#this is for when data is updated on a daily basis online
URL = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv?accessType=DOWNLOAD&bom=true&format=true&delimiter=%3B"
baby_names = pd.read_csv(URL, sep=";")
baby_names.head(3)
```

```python
girls = baby_names[baby_names["Gender"] == "FEMALE"]
boys = baby_names[baby_names["Gender"] == "MALE"]
```

```python
#exporting these boys and girls dataframe to xlsx
#preps the foundation for export as xlsx, a bit more complicated than a regular csv
excel_file = pd.ExcelWriter("Baby Names.xlsx")
```

```python
#use the to_excel parameter to store the dataframe as an excel object
girls.to_excel(excel_file, sheet_name="Girls", index = False)
```

```python
#may use the columns parameter to save only certain columns
boys.to_excel(excel_file, sheet_name="Boys", index = False, columns = ["Year of Birth", "Gender", "Child's First Name"])
```

```python
#MUST use the save the excel foundation object to save
excel_file.save()
```
