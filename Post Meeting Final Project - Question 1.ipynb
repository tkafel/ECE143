```python
#create a new column to know the ratio of evaluations to enrollments
df["student ratio"] = (df["Evals Made"]/df["Enroll"]) *100
```

```python
#group by course to get unique courses
course_group = df.groupby(["Course Number"])
len(course_group)
#create a new dataframe to store all the high ratioed student counts in there
high_evals_made_df = pd.DataFrame(columns=df.columns)
```

```python
for ratio_val, unique_course in course_group:
    highest_valued_ratio = unique_course.nlargest(1, "student ratio")
    high_evals_made_df = high_evals_made_df.append(highest_valued_ratio)
```
