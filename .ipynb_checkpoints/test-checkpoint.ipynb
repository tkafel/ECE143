```python
data = pd.read_csv('Team21_Dataset.csv')
data = data.dropna(axis=0)  # drop NaN
data.reset_index(drop=True, inplace=True)
#data.drop(data.tail(1).index,inplace=True)  # drop last column, which is the same as title, we don't need two
data["Enroll"] = pd.to_numeric(data["Enroll"])  # change enroll into the numeric
data["Evals Made"] = pd.to_numeric(data["Evals Made"])  # change evals_made into the numeric
#print(data["Rcmnd Class"].str.split('%', expand=True)[0])
data["Rcmnd Class"] = pd.to_numeric(data["Rcmnd Class"].str.split('%', expand=True)[0]).div(100)  # change recommended class into the numeric
data["Rcmnd Instr"] = pd.to_numeric(data["Rcmnd Instr"].str.split('%', expand=True)[0]).div(100)  # change recommended Instr into the numeric
data["Study Hrs/wk"] = pd.to_numeric(data["Study Hrs/wk"])  # change hour study into the numeric
data['Avg Grade Expected'] = pd.to_numeric(data['Avg Grade Expected'].str.split('[(|)]', expand=True)[1])  # change avg grade expected into the numeric
data['Avg Grade Received'] = pd.to_numeric(data['Avg Grade Received'].str.split('[(|)]', expand=True)[1])  # change avg grade received into the numeric
# split course description into course number and name
course_split = data['Course'].str.split('-', n=1, expand=True)
data['course_num'] = course_split[0]
data['course_name'] = course_split[1].str.split('[(|)]', expand=True)[0]
data['Term'] = data['Term'].astype('category')
course_split = data['Course'].str.split('-', n=1, expand=True)
dep = course_split[0].str.split(' ', n=1, expand=True)
data['department'] = dep[0]
# find same course and combine
#find_same_course = data[['Instructor','course_num','course_name','Term']].copy().sort_values(by=['course_num'])
#find_same_course.head(40)
data.head(40)
```

```python
combine_data = data.loc[0].copy().to_frame().transpose()
combine_data.iloc[0]['Enroll'] = [combine_data.iloc[0]['Enroll']]
combine_data.iloc[0]['Evals Made'] = [combine_data.iloc[0]['Evals Made']]
combine_data.iloc[0]['Rcmnd Class'] = [combine_data.iloc[0]['Rcmnd Class']]
combine_data.iloc[0]['Rcmnd Instr'] = [combine_data.iloc[0]['Rcmnd Instr']]
combine_data.iloc[0]['Study Hrs/wk'] = [combine_data.iloc[0]['Study Hrs/wk']]
combine_data.iloc[0]['Avg Grade Expected'] = [combine_data.iloc[0]['Avg Grade Expected']]
combine_data.iloc[0]['Avg Grade Received'] = [combine_data.iloc[0]['Avg Grade Received']]

for i in range(1, data.shape[0]):
    #if i / 500 == 1:
    #    break
    #print(data.loc[i]['course_num'] not in combine_data['course_num'] or data.loc[i]['Instructor'] not in combine_data['Instructor'] or data.loc[i]['course_name'] not in combine_data['course_name'])
    check_empty = combine_data.loc[(combine_data['Term'] == data.iloc[i]['Term']) & (combine_data['course_num'] == data.iloc[i]['course_num']) & (combine_data['Instructor'] == data.iloc[i]['Instructor']) & (combine_data['course_name'] == data.iloc[i]['course_name'])]
    if check_empty.empty:
    #if data.iloc[i]['course_num'] not in combine_data['course_num'] or data.iloc[i]['Instructor'] not in combine_data['Instructor'] or data.iloc[i]['course_name'] not in combine_data['course_name']:
        tmp = data.loc[i].copy().to_frame().transpose()
        tmp.iloc[0]['Enroll'] = [tmp.iloc[0]['Enroll']]
        tmp.iloc[0]['Evals Made'] = [tmp.iloc[0]['Evals Made']]
        tmp.iloc[0]['Rcmnd Class'] = [tmp.iloc[0]['Rcmnd Class']]
        tmp.iloc[0]['Rcmnd Instr'] = [tmp.iloc[0]['Rcmnd Instr']]
        tmp.iloc[0]['Study Hrs/wk'] = [tmp.iloc[0]['Study Hrs/wk']]
        tmp.iloc[0]['Avg Grade Expected'] = [tmp.iloc[0]['Avg Grade Expected']]
        tmp.iloc[0]['Avg Grade Received'] = [tmp.iloc[0]['Avg Grade Received']]
        combine_data = combine_data.append(tmp, ignore_index = True)
    #course is the same, combine
    else:
        idx = combine_data[(combine_data['Term'] == data.iloc[i]['Term']) & (combine_data['course_num'] == data.iloc[i]['course_num']) & (combine_data['Instructor'] == data.iloc[i]['Instructor']) & (combine_data['course_name'] == data.iloc[i]['course_name'])].index[0]
        tmp = data.loc[i].copy().to_frame().transpose()
        combine_data.loc[idx]['Enroll'].append(tmp.iloc[0]['Enroll'])
        combine_data.loc[idx]['Evals Made'].append(tmp.iloc[0]['Evals Made'])
        combine_data.loc[idx]['Rcmnd Class'].append(tmp.iloc[0]['Rcmnd Class'])
        combine_data.loc[idx]['Rcmnd Instr'].append(tmp.iloc[0]['Rcmnd Instr'])
        combine_data.loc[idx]['Study Hrs/wk'].append(tmp.iloc[0]['Study Hrs/wk'])
        combine_data.loc[idx]['Avg Grade Expected'].append(tmp.iloc[0]['Avg Grade Expected'])
        combine_data.loc[idx]['Avg Grade Received'].append(tmp.iloc[0]['Avg Grade Received'])
        combine_data.append(tmp)
```

```python
combine_data.sort_values(by=['course_num'])
for i in range(combine_data.shape[0]):
    combine_data.iloc[i]['Enroll'] = sum(combine_data.iloc[i]['Enroll'])
    combine_data.iloc[i]['Evals Made'] = sum(combine_data.iloc[i]['Evals Made'])/len(combine_data.iloc[i]['Evals Made'])
    combine_data.iloc[i]['Rcmnd Class'] = sum(combine_data.iloc[i]['Rcmnd Class'])/len(combine_data.iloc[i]['Rcmnd Class'])
    combine_data.iloc[i]['Rcmnd Instr'] = sum(combine_data.iloc[i]['Rcmnd Instr'])/len(combine_data.iloc[i]['Rcmnd Instr'])
    combine_data.iloc[i]['Study Hrs/wk'] = sum(combine_data.iloc[i]['Study Hrs/wk'])/len(combine_data.iloc[i]['Study Hrs/wk'])
    combine_data.iloc[i]['Avg Grade Expected'] = sum(combine_data.iloc[i]['Avg Grade Expected'])/len(combine_data.iloc[i]['Avg Grade Expected'])
    combine_data.iloc[i]['Avg Grade Received'] = sum(combine_data.iloc[i]['Avg Grade Received'])/len(combine_data.iloc[i]['Avg Grade Received'])
```

```python
term_count = ece.groupby('Term', as_index=False)['Course'].count()
term_count = term_count.rename(columns={"Course":"count"})
ece_fall = term_count.loc[term_count['Term'].str.startswith('FA')].values
ece_winter = term_count.loc[term_count['Term'].str.startswith('WI')].values
ece_spring = term_count.loc[term_count['Term'].str.startswith('SP')].values
school_year = [i for i in range(2007,2019)]
plt.figure(figsize=(10,5))
plt.plot(school_year,ece_fall[:,1],'-bo',label='fall')
plt.plot(school_year,ece_winter[:,1],'-ro',label='winter')
plt.plot(school_year,ece_spring[:,1],'-go',label='spring')
plt.title('ECE fall course over the year')
plt.xlabel('school year')
plt.ylabel('courses number')
plt.legend()
plt.show()
```
